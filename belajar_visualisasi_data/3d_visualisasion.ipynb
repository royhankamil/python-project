untuk menampilkan grafik yang divisualisasikan dalam 3 dimensi

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
# menyiapkan window baru
fig = plt.figure(figsize=(9, 9))

# menentukan jenis grafik itu 3d
ax = plt.axes(projection='3d')
```

menentukan garis dengan tiga titik

```python
x = np.linspace(-6, 6, 100)
y = np.linspace(-6, 6, 100)

x
```

diubah menjadi matriks yang diulang beberapa kali

```python
X, Y = np.meshgrid(x, y)
```

```python
X
```

```python
Y
```

menentukan nilai dari z

```python
z = np.sqrt(X ** 2 + Y ** 2 + 10 )
```

menampilkan hasil 3d

```python
fig = plt.figure(figsize=(8, 8))

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, z)
```

```python
fig = plt.figure(figsize=(8, 8))

ax = plt.axes(projection='3d')

# antialiased itu untuk garis-garis bidang jika di set false maka akan jadi hilang garis nya di object 3d nya
ax.plot_surface(X, Y, z, antialiased=False)
```

mengubah warna / cmap

```python
fig = plt.figure(figsize=(10, 10))

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, z, antialiased=False, cmap='winter')

# untuk menambahkan label pada koordinat x, y, z
ax.set(xlabel='x', ylabel='y', zlabel='z')
```

Cara untuk menentukan view / sudut pandang

```python
fig = plt.figure(figsize=(10, 10))

ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, z, antialiased=False, cmap='winter')

ax.set(xlabel='x', ylabel='y', zlabel='z')

# elev => derajat putaran / rotasi secara vertikal (atas / bawah)
# azim => derajat putaran / rotasi secara horizontal (kanan / kiri)
ax.view_init(elev=-20, azim=45)
```

membuat untuk wireframe 3d (skeleton object)

```python
fig = plt.figure(figsize=(10, 10))

ax = plt.axes(projection='3d')

ax.plot_wireframe(X, Y, z, antialiased=False, cmap='winter')

ax.set(xlabel='x', ylabel='y', zlabel='z')

ax.view_init(elev=-20, azim=45)
```

membuat contour object 

```python
fig = plt.figure(figsize=(10, 10))

ax = plt.axes(projection='3d')

# ada tambahan setelah argumen z, fungsinya untuk menentukan banyaknya lingkaran pada contour
ax.contour3D(X, Y, z, 50, antialiased=False, cmap='winter')

ax.set(xlabel='x', ylabel='y', zlabel='z')

ax.view_init(elev=-20, azim=45)
```
