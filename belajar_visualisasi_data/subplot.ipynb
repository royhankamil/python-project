Apa itu subplot ?
subplot adalah kumpulan plot matplotlib yang digabungkan menjadi satu 
bisa menggabungkan kumpulan histagram, scatter, plot, dll

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=80, cumulative=True); 
```

MATLAB STYLE

```python
plt.subplot(231) # cara membacanya adalah buat subplot 2 baris 3 kolom dan ini adalah urutan subplot ke 1
plt.hist(numbers, bins=80, cumulative=True); 

plt.subplot(233) # cara membacanya adalah buat subplot 2 baris 3 kolom dan ini adalah urutan subplot ke 3
plt.hist(numbers, bins=80, cumulative=True); 

plt.subplot(235) # cara membacanya adalah buat subplot 2 baris 3 kolom dan ini adalah urutan subplot ke 5
x = np.array([i for i in range(50)])
y = np.random.random(50)

plt.scatter(x, y)

plt.subplot(236) # cara membacanya adalah buat subplot 2 baris 3 kolom dan ini adalah urutan subplot ke 6
plt.hist(numbers, bins=80, cumulative=True); 
```

MATPLOTLIB STYLE

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10)); # menyiapkan frame nya diawal dengan 2 baris dan 3 kolom

# meanambahkan plot pada baris 0 dan kolom 0
ax[0, 0].hist(numbers, bins=80, cumulative=True);

# untuk menghilangkan plot koordinat kosong
ax[0, 2].axis('off')

# meanambahkan plot pada baris 0 dan kolom 1
ax[0, 1].hist(numbers, bins=80, cumulative=True);

# meanambahkan plot pada baris 1 dan kolom 1
ax[1, 1].scatter(x, y);

# meanambahkan plot pada baris 1 dan kolom 2
ax[1, 2].hist(numbers, bins=100, density=1);
```

cara untuk menyamakan titik koordinat x dan y nya

```python
fig, ax = plt.subplots(2, 3, figsize=(15, 10), sharex=True, sharey=True); # menyiapkan frame nya diawal dengan 2 baris dan 3 kolom

# meanambahkan plot pada baris 0 dan kolom 0
ax[0, 0].hist(numbers, bins=80, cumulative=True);

# untuk menghilangkan plot koordinat kosong
ax[0, 2].axis('off')

# meanambahkan plot pada baris 0 dan kolom 1
ax[0, 1].hist(numbers, bins=80, cumulative=True);

# meanambahkan plot pada baris 1 dan kolom 1
ax[1, 1].scatter(x, y);

# meanambahkan plot pada baris 1 dan kolom 2
ax[1, 2].hist(numbers, bins=100, density=1);
```
