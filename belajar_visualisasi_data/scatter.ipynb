kali ini kita akan membuat sebuah histogram

```python
import matplotlib.pyplot as plt
import numpy as np
```

bins adalah pemisah nya ya misal data bins nya = 25 

maka akan dibagi menjadi 25 bagian yang mana dimulai dari data terkecil hingga terbesar (secara rata)

yang mana hasil dari bagian itu akan menjadi titik x dari histagram

```python
numbers = 3 * np.random.rand(13)
print(numbers)

plt.hist(numbers, bins=20)
```

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=80); # titik koma untuk menghilangkan text dari hasil list bins
```

density mirip seperti probability jadi nilai dari y yang sebelumnya 0 - 400 akan dirubah menjadi 0 - 1

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=80, density=1);
```

bisa juga ditambahkan cumulative yakni diurutkan dari yang terkcil hingga terbesar

```python
numbers = np.random.randn(10000)
plt.hist(numbers, bins=80, cumulative=True); 
```
