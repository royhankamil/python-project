Bagaimana caranya untuk menambahkan text di plot pada matplotlib ?

Dan bagaimana caranya untuk menambahkan anotasi di plot matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
numbers = np.random.random(30)
plt.figure(figsize=(6, 6))
plt.hist(numbers, bins=80, cumulative=True)

# cara untuk menampilkan garis lurus, parameter x = 0.5
plt.axvline(0.5, color='r')

# cara untuk menampilkan text di matplotlib, parameter x = 0.5, y = 0.7, isi_text
plt.text(0.5, 20, "textnya")

# cara untuk menampilkan text dengan horizontal align center ha => horizontal allign
plt.text(0.5, 15, "tengah", ha='center')


# cara untuk menampilkan anotasi, parameter => text, posisi_x_y, posisi_text, allign_text, jenis_arrow 
plt.annotate("tengah", xy=(0.5, 20), xytext=(0.4, 25), ha='center', arrowprops={'arrowstyle': '-|>'})
```
