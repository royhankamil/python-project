Latihan untuk belajar pandas membuat filtering

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("insurance.csv")

df.head()
```

```python
df.info()
```

melakukan compare / perbandingan antar data
contoh : apakah data pada kolom age yang lebih besar dari 20

```python
df.age > 20 # menamplkan apakah true / false pada setiap baris

df[df.age > 20] # menampilkan data yang age > 20 saja
```

menampilkan data yang mempunyai anak saja

```python
df[~(df.children == 0)]
```

multiple statement

```python
df[~(df.children == 0) & (df.age < 20)]
```

```python
df[(df.region.isin(["northwest", "southeast"]) | (~(df.children == 0) & (df.age < 20)))]
```

tetapi ingat itu hanya menampilkan datanya saja untuk menyimpannya menjadi dataframe baru maka harus disimpan di variable baru dan ditambahkan di belakangnya .copy() untuk menjaga dataframe yang asli

Melakukan sorting

ascending => melakukan pengurutan dari bawah ke atas

```python
df.sort_values("age", ascending = True)
```

melakukan groupby -> pemisahan antar baris data berdasarkan kelompok tertentu
lalu hasil penggabungannya bisa berasal mean(), median(), atau count()

```python
numerik_df = df.drop(columns=["sex", "smoker"]).copy()

numerik_df.groupby("region").mean()
```

dengan menggunakan 2 tipe kategorikal

```python
new_numerik_df = df.drop(columns=["smoker"]).copy()

new_numerik_df.groupby(["region", "sex"]).median()
```

melakukan pivot table yakni menentukan table baris, kolom, values berdasarkan kolom data sebelumnya

```python
df.pivot_table(index= "region", columns="sex", values="bmi", aggfunc="mean")
```

membuat kolom baru dan menghilangkan kolom

```python
df["new_columns"] = np.random.rand(len(df))
df.head()
```

```python
df.drop(columns="new_columns", inplace=True)
df.head()
```

Map, Apply, dan Transform

map, yakni membuat value kolom sesuai value object / kategori kolom lain

```python
df["category_region"] = df.region.map({'southwest': 1, 'southeast':2, 'northwest':3, 'northeast': 4})
df.head()
```

membuat value kolom sesuai dengan kolom lain yang sudah dimasukkan di dalam function

```python
def convert_yes_no_to_numerik(x):
    if x == "yes":
        return 1
    else:
        return 0
```

```python
df["num_smoker"] = df.smoker.apply(convert_yes_no_to_numerik)

df.head()
```

apply juga dapat dimasukkan lambda function

```python
df["bmi_kuadrat"] = df.bmi.apply(lambda x: x*x)
df.head()
```

transform yakni pengubahan x menjadi numerik tertentu misal kita ingin menjadikan value age baru menjadi value 0-1

```python
df['scaled_age'] = df.age.transform(lambda x: (x-x.min())/(x.max()-x.min()))
df.head()
```
